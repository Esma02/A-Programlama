```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Gelenler'], kde=True)
plt.title('Gelenler Dağılımı')
plt.xlabel('Gelenler')
plt.ylabel('Frekans')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Gelenler'])
plt.title('Gelenler Kutu Grafiği')
plt.ylabel('Gelenler')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='İl Sıra', y='Gelenler')
plt.title('İl Sırası ve Gelenler Arasındaki İlişki')
plt.xlabel('İl Sırası')
plt.ylabel('Gelenler')
plt.show()
```

```python
gidenler_sutunu = df['Gelenler']
Q1 = np.percentile(gidenler_sutunu, 25)
Q3 = np.percentile(gidenler_sutunu, 75)
ceyrekler_acikligi = Q3 - Q1
print("Çeyrekler Açıklığı:", ceyrekler_acikligi)
```
